```python
# Model training utilities
from sklearn.svm import LinearSVC
import joblib
```

```python
# Load preprocessed and vectorized training data
X_train_vec, X_test_vec, y_train, y_test = joblib.load(
    "data/processed/vectorized_data.pkl"
)
```

```python
# Initialize and train a Linear Support Vector Machine classifier
svc = LinearSVC()
svc.fit(X_train_vec, y_train)
```

```python
# Save trained model for evaluation and future inference
joblib.dump(svc, "data/processed/svm_model.pkl")
```
