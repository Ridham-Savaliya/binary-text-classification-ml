```python
# Libraries for feature extraction and data splitting
import pandas as pd
from sklearn.model_selection import train_test_split
from sklearn.feature_extraction.text import TfidfVectorizer
import joblib
```

```python
# Load cleaned text data
df = pd.read_csv("IMDB Dataset.csv")

```

```python
# Split data into training and testing sets
# Important: split BEFORE vectorization to prevent data leakage
X_train, X_test, y_train, y_test = train_test_split(
    df['review'],
    df['sentiment'],
    test_size=0.2,
    random_state=42
)
```

```python
# Convert text data into numerical features using TF-IDF
vectorizer = TfidfVectorizer(max_features=5000)

X_train_vec = vectorizer.fit_transform(X_train)
X_test_vec = vectorizer.transform(X_test)
```

```python
# Persist vectorizer and transformed data for reuse in training and evaluation
joblib.dump(vectorizer, "data/processed/tfidf_vectorizer.pkl")
joblib.dump(
    (X_train_vec, X_test_vec, y_train, y_test),
    "data/processed/vectorized_data.pkl"
)
```
